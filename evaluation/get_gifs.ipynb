```python
exp = []

temps = [0.5, 1, 2]
emb_lens = [16, 32, 64]

for temp in temps:
    for emb_len in emb_lens:
        for model_name in ["Linear"]:
            c = {}
            c["task"] = "reg"
            c["model_name"] = model_name
            c["aux_feats"] = []
            
            c["model_params"] = {
                "emb_len": 16,
                "temperature": 1,
                "use_emb": False,
            }
            exp.append(c)
            
            c = {}
            c["task"] = "reg"
            c["model_name"] = model_name
            c["aux_feats"] = []
            
            c["model_params"] = {
                "emb_len": emb_len,
                "temperature": temp,
                "use_emb": True,
            }
            exp.append(c)
exp
```

```python
gen_gif(out_path="./gifs/linaer_emb_T.gif", paths=paths, target_filter=exp,
        name_parse=lambda x: x["model_name"] + " N=" + str(x["model_params"]["emb_len"]) + " T=" + str(x["model_params"]["temperature"]),
        sort_name=lambda x: list(sorted(x, key=lambda x: [float(y.split("=")[1]) for y in x.split(' ')[1:]])))
```
